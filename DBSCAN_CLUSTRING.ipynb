```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import DBSCAN
from sklearn.datasets import make_blobs #blobs-it creates clusters of data points that look like clouds
from sklearn.preprocessing import StandardScaler #Feature Scaling

# 1. Generate Data
X, y = make_blobs(n_samples=500, centers=5, cluster_std=0.7, random_state=42)

# 2. Scale Data
A = StandardScaler()
X_NEW = A.fit_transform(X)

# --- NEW CODE: Visualize Raw Data ---
plt.figure(figsize=(12, 5))

# Subplot 1: The Raw Dataset
plt.subplot(1, 2, 2) # for grid row, column ,index
plt.scatter(X_NEW[:, 0], X_NEW[:, 1], c='gray', alpha=0.6)
plt.title("Original Scaled Dataset (No Labels)")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")

# 3. Apply DBSCAN
dbscan = DBSCAN(eps=0.8, min_samples=18)
labels = dbscan.fit_predict(X_NEW)

# Subplot 2: The Clustered Data
plt.subplot(1, 2, 1)
plt.scatter(X_NEW[:, 0], X_NEW[:, 1], c=labels, cmap='viridis')
plt.title("DBSCAN Clustering Results")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")


plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import DBSCAN
from sklearn.datasets import make_blobs #blobs-it creates clusters of data points that look like clouds
from sklearn.preprocessing import StandardScaler #Feature Scaling

# 1. Generate Data
X_data, _ = make_blobs(n_samples=300, centers=3, cluster_std=0.5, random_state=42)

# 2. Scale Data
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X_data)

# --- NEW CODE: Visualize Raw Data ---
plt.figure(figsize=(10, 4))

# Subplot 1: The Raw Dataset
plt.subplot(1, 2, 1) # for grid row, column ,index
plt.scatter(X_scaled[:, 0], X_scaled[:, 1], c='gray', alpha=0.6)
plt.title("Original Scaled Dataset (No Labels)")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")

# 3. Apply DBSCAN
dbscan = DBSCAN(eps=0.8, min_samples=5)
labels = dbscan.fit_predict(X_scaled)

# Subplot 2: The Clustered Data
plt.subplot(1, 2, 2)
plt.scatter(X_scaled[:, 0], X_scaled[:, 1], c=labels, cmap='plasma')
plt.title("DBSCAN Clustering Results")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")

plt.tight_layout()
plt.show()
```
